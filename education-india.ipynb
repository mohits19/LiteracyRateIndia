## Preparing for Regression

By performing regression, we can look at how each variables affects our target variable, *OVERALL_LI* (Overall Literacy)

```python
X = data.drop(['STATNAME','DISTNAME','OVERALL_LI','FEMALE_LIT','MALE_LIT'],axis=1)
y = data['OVERALL_LI']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = pd.DataFrame(sc.fit_transform(X),columns=X.columns,index=X.index)
```

```python
from sklearn.linear_model import LinearRegression
lm0 = LinearRegression()
lm0.fit(X_scaled,y)
```

```python
ypred0 = lm0.predict(X_scaled)
resid0 = y - ypred0
```

```python
sns.distplot(resid0)
plt.show()
```

```python
from scipy import stats
stats.shapiro(resid0)
```

- pvalue < 0.05 implies we reject null hypothesis
- residuals are NOT normally distributed

```python
resid0.mean()
```

- Mean of residuals is approximately zero

## Regression

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(resid0, X_scaled)
lzip(name, test)
```

- pvalue > 0.05 implies we fail to reject null hypothesis
- residuals are homoscedastic

```python
from statsmodels.stats import diagnostic as diag
min(diag.acorr_ljungbox(resid0, lags=628)[1])
```

- pvalue > 0.05 implies we cannot reject $H_0$
- residuals are not autocorrelated

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=1)
```

```python
from statsmodels.api import add_constant
X_const = add_constant(X_train)
```

```python
import statsmodels.formula.api as sm
lm1 = sm.OLS(y_train,X_const).fit()
lm1.params
```

```python
print(lm1.summary())
```

```python
imp_cols = lm1.pvalues[lm1.pvalues < 0.05].index
imp_cols
```

```python
pd.DataFrame({'Coeff': lm1.params[imp_cols], 'P-value':lm1.pvalues[imp_cols]})
```

```python
from sklearn.metrics import r2_score
lm2 = LinearRegression()
lm2.fit(X_train,y_train)

y_pred = lm2.predict(X_train)
print('Train R^2 =',r2_score(y_pred,y_train))

y_pred = lm2.predict(X_test)
print('Test R^2 =',r2_score(y_pred,y_test))
```

## Comparison of Algorithms

```python
models = [('Linear Regression', LinearRegression()),
         ('Ridge', Ridge(alpha=43.0)),
         ('KNN', KNeighborsRegressor(metric='manhattan', n_neighbors=3)),
         ('DecisionTree', DecisionTreeRegressor(criterion='mae', max_depth=5, max_leaf_nodes=10, min_samples_leaf=5, min_samples_split=2)),
         ('RandomForest', RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=4, max_features='auto')),
         ('AdaBoost', AdaBoostRegressor(learning_rate=0.01)),
         ('GradientBoost', GradientBoostingRegressor(learning_rate=0.01))]
```

```python
seed = 7
results = []
names = []
```

```python
from sklearn.model_selection import KFold, cross_val_score
for name, model in models:
    kfold = KFold(n_splits=10, random_state=seed)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='r2')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
fig, ax = plt.subplots(figsize=(14,6))
plt.boxplot(results)
plt.title('Algorithm Comparison')
ax.set_xticklabels(names)
plt.ylabel('Rsq')
plt.savefig('graphs/box.png')
```
